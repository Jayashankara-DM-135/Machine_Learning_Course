###### Aggregate statistics (Groupby)

```python
df = pd.read_csv('modified.csv')
print(df)

#Group all the pokemans based on Type 1 and list which is having more attack
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)

#Group all the pokemans based on name and list which is having more HP score.

df.groupby(['Name']).mean().sort_values('HP', ascending=False)

#Group all the pokemans and list out sum of HP for each group

#df.groupby(['Name']).sum()



#Group pokeman based on type 1 and count how many of them in each group.

print("++++++++++++++++++++++++++++")
print(df)
#Below line just to assign each occurence as 1, so thta we can use these colums to count.
df['count'] = 1
print(df)
print("=======================")
df.groupby(['Type 1']).count()['count']

#We can also group by two fileds 
df.groupby(['Type 1', 'Type 2']).count()['count']
```

# How to read a huge file 

```python
#Basically we are creating new data frame by using columns name of alreday existing data frame called 'df'
# But it will not contains no rows.

new_df = pd.DataFrame(columns=df.columns)
print(new_df)

for df in pd.read_csv('modified.csv', chunksize=5):
    res = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, res])

print(new_df)
```
