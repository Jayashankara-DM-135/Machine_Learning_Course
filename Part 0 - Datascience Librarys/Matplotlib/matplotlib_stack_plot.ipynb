```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

minites= [1, 2, 3, 4, 5]
player1 = [0, 1, 1, 2, 3]
player2 = [1, 1, 2, 3, 4]
player3 = [0, 0, 1, 2, 3]

labels = ['player1', 'player2', 'player3']

plt.stackplot(minites, player1, player2, player3, labels=labels)
plt.tight_layout()
plt.legend(loc='upper left')
plt.show()
```
