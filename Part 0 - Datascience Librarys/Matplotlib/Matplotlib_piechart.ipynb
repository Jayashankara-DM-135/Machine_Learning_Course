```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
slices = [60, 40]
plt.pie(slices)
plt.title("first Pie chart")
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

slices = [17, 196]
labels = ['17%', '196%']

plt.pie(slices, labels=labels)
plt.title("Second Pie chart with label")
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')
slices = [20, 30, 30, 10]
labels = ["RCB", "CSK", "MI", "RR"]
colors = ['red', 'yellow', 'blue', 'pink']
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'b'})
plt.title("Thrid Pie chart with marker")
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

slices = [10, 29, 17, 35, 15]
labels = ['C', 'Java', 'C#', 'Python', 'C++']

#Explod python programing language.
"""
0 ==> Means nothing
anything other than 0 is gap ratio.

Postive value creates the space 
"""
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('Programing Language')
plt.tight_layout()
plt.show()
```
